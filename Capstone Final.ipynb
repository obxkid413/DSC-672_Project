```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
# Draw a catplot to show average salary distributon for year and gender
from numpy import median, mean
g = sns.catplot(x="fy", y="salary", hue="gender", data=completenew,
                height=6, kind="bar", palette="muted", estimator=mean)
g.despine(left=True)
g.set_ylabels("salary")
```

```python
#Show boxplot of salaries by zipcode.
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)
g = sns.boxenplot(x="race_ethnicity_desc", y="salary", data=completenew)
g.set_xticklabels(g.get_xticklabels(), rotation=25)
```
